**Gulls- inhabits subarctic and temperate regions of the Northern Hemisphere, Laughing gulls are most likely to strike aircraft between May and September, with the highest number of strikes occurring in June and July


**White tailed deer- covers a huge range from the Arctic Circle in western Canada to 18 degrees south of the Equator in Peru and Bolivia. Its body and antlers are largest in cold temperate climates and on productive agricultural soils but are small in the tropics, in deserts, and on small islands. 

*The mating season for white-tailed deer, also known as the rut, takes place in late October to early December. During this time, deer are more active and are more likely to collide with vehicles, especially in November. The rut lasts about three weeks in the northern states, and the timing is influenced by the moon's phases*


**Canada goose - migrates twice a year. In September or October, Canada geese fly south to their non-breeding sites to avoid the cold. They stay in their non-breeding sites all winter. Geese migrate north to their breeding sites in April, May or Early June


**OVERALL**

**Bird strikes are a common threat to aircraft safety, but they don't pose a significant risk overall. To avoid bird strikes, pilots can:
*Avoid areas where birds congregate, such as marshlands and landfills
*Avoid flying beneath flocks of birds
*Pitch up when approaching a bird

**Bird strikes are the greatest potential hazard to aircraft because of their size, abundance, or habit of flying in dense flocks. This time of year, airplanes are at a much greater risk to bird strikes by Canada geese.

**These bird strikes increase during the migration months of March, April, August, September, October and November as revealed in our data. Research shows that the altitudes of migrating Canada geese vary with winds aloft, weather fronts, terrain elevations, cloud conditions, and other environmental variables.  



<div class="alert alert-info" style="background-color:#006a79; color:white; padding:0px 10px; border-radius:5px;"><h2 style='margin:10px 5px'>RECOMMENDATION</h2>
</div>

**While over 90% of reported bird strikes occur at or below 3,000 feet, strikes at higher altitudes are common during migration, hence, Pilots are cautioned to minimize en route flying at lower altitudes during migration as it may pose risk or danger tothe aircraft and passengers.

**Bird strike mitigation
Methods to mitigate bird strikes include removing nests from trees, developing devices to scare birds, and removing swamps around airports

**Checking notams for bird activity near airports
Checking the FAA Airport/Facility Directory (A/FD) for warnings regarding bird hazards
Turning lights on when flying in an area with flocks of birds
